# Práctica de Web Scraping 

## Importar librerías

Importo la librería [requests](https://docs.python-requests.org/en/latest/). Usada para facilitar las peticiones HTTP para Python.

```python
import requests
```

Voy a importar de la librería [bs4](https://www.crummy.com/software/BeautifulSoup/bs4/doc/) `BeautifulSoup`. Permite extraer información en formato html

```python
from bs4 import BeautifulSoup
```

## Variables

Es un nombre que hace referencia a un valor determinado, pudiendo ser un string, un número entero...

### Definimos URL

A la variable url le asignamos el valor del enlace al medallero de los juegos olimpicos.

```python
 URL = "https://resultados.elpais.com/deportivos/juegos-olimpicos/medallero/"
```

### Realizamos la petición a la web

Realizamos una petición a la web del medallero de los JJOO y la guardamos en la variable req. 

Si el estatus code de la petición no es `200` no se puede leer la página. Para leer la página el estatus code (Resultado del estado de la petición) que necesitamos tiene que ser obligatoriamente 200.


```python
req = requests.get(URL)
if (req.status_code != 200):
    raise Exception("No se puede hacer Web Scraping en"+ URL)
if (req.status_code == 200):
    print("Vamos a por ello")
```

### De requests a BeautifulSoup

La librería BeautifulSoup nos permite traducir el código importado anteriormente a un lenguaje html, en este caso. El resultado de este proceso lo guardamos en la variable html.


```python
html = BeautifulSoup(req.text, "html.parser")
```

### Variables de datos

Definimos las variables `paises`, `oros`, etc para asociar la información encontrada en el fichero teniendo en cuenta los filtros que hemos establecido a la función find_all.

```python
paises = html.find_all("th",{"class":"pais"})
oros = html.find_all("td",{"class":"m_oro"})
platas = html.find_all("td",{"class":"m_plata"})
bronces = html.find_all("td",{"class":"m_bronce"})
totales = html.find_all("td",{"class":"m_total"})
```

## Hacemos la pregunta

Con la funcion input le preguntamos al usuario si quiere conocer los paises que han obtenido medallas y esperamos su respuesta, que tiene que ser s si los quiere conocer, este valor lo guardamos en la variable respuesta. Comprobamos que el valor de la variable respuesta es una s para mostrar los resultados.

```python
respuesta=input('¿QUIERES CONOCER LOS 20 PAÍSES QUE HAN OBTENIDO MÁS MEDALLAS EN 2020?\n \n Si tu respuesta es Sí, presiona "s" \n')
if(respuesta == 's'): 
    print('Vale, vamos a ello')
```

## Bucle para obtener los datos

Mostramos en pantalla un mensaje a los usuarios mostrando los resultados de las medallas. Gracias al for que nos recorre los primeros 20 términos, en este caso, los primeros 20 países, vamos a ir imprimiendo por pantalla las medallas de oro, plata y bronce de cada país mencionado anteriormente.

```python
print('\nRESULTADOS DE LOS DATOS DE LOS JUEGOS OLÍMPICOS 2020\n')
print ('PAÍSES')
print(paises)
for i in range (20):
# Con el método "getText()" no nos devuelve el HTML
    print("%d. %s \nOro: %s Plata: %s Bronce: %s \n Total: %s \n " % (i+1, paises[i+1].text.strip(),oros[i].text.strip(),platas[i].text.strip(),bronces[i].text.strip(), totales[i].text.strip()))
else:
    print('Qué lástima, y...')
```
